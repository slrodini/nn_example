```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
ax = plt.figure().add_subplot(projection='3d')
for i in range(0,11):
    data = np.loadtxt("res_t_"+str(i)+".dat")
    print("loaded :res_t_"+str(i)+".dat")
    ax.plot(data[:,0],data[:,1],11-i,zdir='x')
```
